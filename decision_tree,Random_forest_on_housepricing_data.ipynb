```python
import pandas as pd 
import numpy as np
house = pd.read_csv('housebuilding.csv') #reading the data
```

```python
"""treating missing values , decision tree and random forest deal with missing value on their own
but here i treated missing values by myself"""
house.dropna(thresh=1430,inplace=True,axis=1) #threshold is set to keep columns having at least 1430 values
house.isnull().sum()
```

```python
#treating some left out missing values by using 'fillna'
house.fillna({'MasVnrType':'BrkFace','MasVnrArea':np.mean(house['MasVnrArea'])},inplace=True)
house.fillna('SBrkr',inplace=True)
house.isnull().sum()
```

```python
#splitting our dataset
from sklearn.model_selection import train_test_split
y = house['SalePrice']
house.drop(columns='SalePrice',inplace=True)

X_train,X_test,y_train,y_test = train_test_split(house,y,test_size=0.2,random_state=42)
y_train
```

```python
#doing label encoding on X_train categorical columns, converting them to numerical values
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for col in X_train.columns.values:

       if X_train[col].dtypes=='object':
            data=X_train[col].append(X_train[col])
            le.fit(data.values)
            X_train[col]=le.transform(X_train[col])
X_train
```

```python
#label encoding the categorical columns present in our test data
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for col in X_test.columns.values:

       if X_test[col].dtypes=='object':
            data=X_test[col].append(X_test[col])
            le.fit(data.values)
            X_test[col]=le.transform(X_test[col])
X_test
```

```python
#applying decision tree regressor machine learning algorithm without any hyperparametric tuning
from sklearn.tree import DecisionTreeRegressor
decision = DecisionTreeRegressor()
decision.fit(X_train,y_train)
predicted = decision.predict(X_test)
predicted
```

```python
#comparing actual and predicted values by creating a dataframe
comparison_df = pd.DataFrame({'actual':y_test,'predicted':predicted})
comparison_df
```

```python
#checking score of our model
from sklearn.metrics import r2_score
r2_score(y_test,predicted)
```

```python
#applying Random Forest model on our data with hyperparametric tuning of only fixing the number of trees we are using to 80
from sklearn.ensemble import RandomForestRegressor
random_f = RandomForestRegressor(n_estimators=80,random_state=42)
random_f.fit(X_train,y_train)
rf_prediction = random_f.predict(X_test)
rf_prediction
```

```python
# checkinh R2 score to check performance of our model on our test data we get 88.6%
r2_score(y_test,rf_prediction)
```
